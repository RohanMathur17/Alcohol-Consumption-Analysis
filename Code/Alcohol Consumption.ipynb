```python

value_counts_fam_relationship=df_por['Family_Relationship'].value_counts().tolist() 
unique_fam_relationship=df_por['Family_Relationship'].unique().tolist() 

perc_fam_relationship = [i/len(df_por) for i in value_counts_fam_relationship ]
perc_fam_relationship
dict_fam_relationship = dict(zip(unique_fam_relationship,perc_fam_relationship))
print(dict_fam_relationship)




value_counts_fam_relationship=df_alcohol['Family_Relationship'].value_counts().tolist() 
unique_fam_relationship=df_alcohol['Family_Relationship'].unique().tolist() 

perc_fam_relationship_alcohol = [i/len(df_por) for i in value_counts_fam_relationship ]
perc_fam_relationship_alcohol
dict_fam_relationship_alcohol = dict(zip(unique_fam_relationship,perc_fam_relationship_alcohol))
print(dict_fam_relationship_alcohol)

```

```python
fig, axs = plt.subplots(1,2 , figsize=(15, 5))
sns.barplot(x=unique_fam_relationship,y = perc_fam_relationship,ax = axs[0] )
sns.barplot( x=unique_fam_relationship ,y=perc_fam_relationship_alcohol,ax=axs[1])
```
