Distribution for the marks of G1 is fairly a normal distribution & happens to be a little left skewed.

Distribution for the marks of G2 is fairly a normal distribution as well & left skewed. However there is a rise in the amount of 
students who got zero.

Distribution for the marks of G3 is a bimodal distribution. The peaks are at 0 & 12 approximately.
